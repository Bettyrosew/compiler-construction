**Lab 4: Understanding Predictive Parsing**

**Using an example or illustration, explain what is predictive parsing?**

Predictive Parsing is an effective technique of executing recursive-descent parsing by managing the stack of activation records.
Predictive Parsers has the following components ‚àí
1.   Input Buffer ‚àí The input buffer includes the string to be parsed followed by an end marker $ to denote the end of the string.

2.   Stack ‚àí It contains a combination of grammar symbols with dollar sign on the bottom of the stack. At the start of Parsing, the stack contains the start symbol of Grammar followed by $.

3.  Parsing Table ‚àí It is a two-dimensional array or Matrix M [A, a] where A is nonterminal and 'a' is a terminal symbol. 

4.  Parsing Program ‚àí The parsing program performs some action by comparing the symbol on top of the stack and the current input symbol to be read on the input buffer. 

5.  Actions ‚àí Parsing program takes various actions depending upon the symbol on the top of the stack and the current input symbol.











**Algorithm to construct Predictive Parsing Table**

Input ‚àí Context-Free Grammar G

Output ‚àí Predictive Parsing Table M

Method ‚àí For the production A ‚Üí Œ± of Grammar G.

* For each terminal, a in FIRST (ùõº) add A ‚Üí Œ± to M [A, a].
* If Œµ is in FIRST (Œ±), and b is in FOLLOW (A), then add A ‚Üí Œ± to M[A, b].
* If Œµ is in FIRST (Œ±), and $ is in FOLLOW (A), then add A ‚Üí Œ± to M[A, $$].
* All remaining entries in Table M are errors.

**What is LL (1) parsing?**


LL(1) parsing is a top-down parsing method in the syntax analysis phase of compiler design.  Required components for LL(1) parsing are input string, a stack, parsing table for given grammar, and parser.
